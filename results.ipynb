```python
import cv2
import matplotlib.pyplot as plt

def show_detections(image_path, boxes, confs=None):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    for i, (x1, y1, x2, y2) in enumerate(boxes):
        color = (0, 255, 0) 
        thickness = 2
        cv2.rectangle(img, (x1, y1), (x2, y2), color, thickness)
        if confs:
            cv2.putText(img,
                        f"{confs[i]:.2f}",
                        (x1, y1 - 5),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.5,
                        color,
                        1)

    plt.figure(figsize=(8,8))
    plt.imshow(img)
    plt.axis('off')
    plt.show()

boxes = [(27, 274, 462, 489)]
confs = [0.12, 0.11]
show_detections('images/outputs/Miami_ (25.798387, -80.196265)_boxes.png', boxes)
```
